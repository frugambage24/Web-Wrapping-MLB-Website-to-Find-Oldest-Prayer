```python
def between(S, u, v):
    """Return a list of substrings in S between occurrences of u and v"""
    result = []
    start_index = 0

    while True:
        start_pos = S.find(u, start_index)
        if start_pos == -1:
            break

        end_pos = S.find(v, start_pos + len(u))
        if end_pos == -1:
            break

        substring = S[start_pos + len(u):end_pos]
        result.append(substring)

        start_index = end_pos + len(v)

    return result
```

```python
class Date:
    def __init__(self, date_string):
        # Split the date_string into month, day, and year
        self.month, self.day, self.year = map(int, date_string.split('/'))

    def __repr__(self):
        return f"{self.month}/{self.day}/{self.year}"

    def nextyear(self):
        return Date(f"{self.month}/{self.day}/{self.year + 1}")

    def inNyears(self, n):
        return Date(f"{self.month}/{self.day}/{self.year + n}")

    def __lt__(self, other):
        # Use the '<' operator to check if one date is before another
        if self.year < other.year or (self.year == other.year and self.month < other.month) or (
                self.year == other.year and self.month == other.month and self.day < other.day):
            return True
        else:
            return False
```

```python
def SeniorPlayer(url):

  roster_url = url
  page=get(roster_url).text
  links = between(page, '<a href="', '"')
  player_links = [link for link in links if contains(link, '/player/')]

  player_debut={}
  for p in player_links:
    #Iterate through each item in player_links, get HTML by adding item to url
    page=get('https://www.mlb.com'+p).text
    FullName=between(page,'<li class="full-name"><span class="label">Fullname:</span> ',"</li>")
    if FullName:
      Debut=between(page,'<li><span class="label">Debut:</span> ',"</li>")
      player_debut[FullName[0]]=Debut[0]

  #Gather all dates
  all_dates=list(player_debut.values())

  #Convert each date into a Date object
  for i in range(len(all_dates)):
    all_dates[i]=Date(all_dates[i])

  #Use first date to determine the earliest date
  earliest_date = FirstDate(all_dates)

  #Return the key (name) of the player with the earliest date

  for fullName, Debut in player_debut.items():
    if str(Date(Debut)) == str(earliest_date):
      return fullName

SeniorPlayer("https://www.mlb.com/angels/roster")
```
